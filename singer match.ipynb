# Choose mode and name singer(s)



```python
mode = input("Do you want to IDENTIFY or LIST or LEARN or GUESS singers?").strip().lower()
if mode == "IDENTIFY".lower():

    gender = input("What gender is the singer").strip().lower()
    age = int(input("How old is the singer?").strip())
    music_type = input("What type of music does the singer make?").strip().lower()
    language = input("What language does your singer make music in?").strip().lower()
    awards = int(input("How many grammys has your singer won?").strip())
    famous_song = input("Enter the artist's most famous song").strip().lower()
    user_list = [gender,age,music_type,language,awards,famous_song]
    
    def check_match(user_list,celebrity_list):
        for name,profile in celebrity_list.items():
           if list(map(str,profile[:6])) == list(map(str,user_list)):
                return name
        return None
    
    singer = check_match(user_list,celebrity_list)
    if singer:
        print(f"your singer is {singer.title()}")
    else:
        print("no match found")
    
    
elif mode == "LIST".lower():
    category_inp = input("name a category input and ill name all the singers from it")
    def singer_from_category(celebrity_list,category):
        matching_singers = []
        for singer,features in celebrity_list.items():
            if category in features:
                matching_singers.append(singer)
        return matching_singers
        
    singers_with_category = singer_from_category(celebrity_list,category_inp)
    if singers_with_category:
        print(f"the singers with matching category {category_inp} are: \n{singers_with_category}")
    else:
        print("No matches")

elif mode == "LEARN".lower():
    learn = input("what singer do you want to learn about?")
    print(f"info about {learn} is provided below in order of \n[gender, age , music_type , language , no. of grammys ,famous_song, hair color, eye color ]:\n")
    print(celebrity_list[learn.title()])

elif mode == "GUESS".lower():
    qual1 = input("tell me one physical quality about your singer(hair)")
    qual2 = input("tell me one more physical quality about your singer(eyes)")
    qual3 = input("tell me your singer's music type")
    qual4 = input("tell me your singer's gender")
    qual5 = int(input("tell me the singer's age"))
    qualities = [qual4,qual3,qual1,qual2,qual5]
    def similarity(a,b):
        return SequenceMatcher(None,a,b).ratio()
    def find_closest_match(qualities):
        max_similarity = 0
        closest_match = None

        
        for singer, attributes in celebrity_list.items():
            # Calculate similarity score between user input and singer's attributes
            sim_score = similarity(qualities, attributes)

            
            if sim_score > max_similarity:
                max_similarity = sim_score
                closest_match = singer
        
        if closest_match:
            print(f"closest match is {closest_match}")
        else:
            print("No matches found :(")
    find_closest_match(qualities)


```
